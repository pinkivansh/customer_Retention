```python
# categorical_columns can be located from column index 0 to 16
categorical_columns=customer_Retention_df.iloc[:,:17].columns
categorical_columns=categorical_columns.to_list()

# rating_columns can be located from column index 17 to 46
rating_columns=customer_Retention_df.iloc[:,17:47].columns
rating_columns=rating_columns.to_list()

# ecommerce_columns can be located from column index 47 and above
ecommerce_columns=customer_Retention_df.iloc[:,47:].columns
ecommerce_columns=ecommerce_columns.to_list()
```

```python
print(customer_Retention_df['1Gender of respondent'].value_counts())
sns.countplot(x=customer_Retention_df['1Gender of respondent'])
plt.title('Gender')
```

```python
print(customer_Retention_df['2 How old are you? '].value_counts())
plt.figure(figsize=(10,4))
sns.countplot(x=customer_Retention_df['2 How old are you? '],hue = customer_Retention_df['1Gender of respondent'])
plt.title('Age')
```

```python
print(customer_Retention_df['3 Which city do you shop online from?'].value_counts())
plt.figure(figsize=(16,4))
sns.countplot(x=customer_Retention_df['3 Which city do you shop online from?'],hue=customer_Retention_df['1Gender of respondent'])
plt.title('City')
```

```python
#Let's see the plot for all the categorical parameters

for col in categorical_columns:
    if col!= '1Gender of respondent' and col!='4 What is the Pin Code of where you shop online from?':  #Already plotted for Gender
        print(customer_Retention_df[col].value_counts())
        plt.figure(figsize=(10,4))
        sns.countplot(x=customer_Retention_df[col])
        plt.show()
        print("-----"*25)
```

```python
#Let's see the counts for pin codes
plt.figure(figsize=(20,4))
sns.countplot(x=customer_Retention_df['4 What is the Pin Code of where you shop online from?'])
plt.xticks(rotation=45)
plt.show()
```
